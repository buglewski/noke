```python
import numpy as np
import random
```

```python
# Play with the normal distribution of c
m = 10

Ec = 1
Dc = 0.25

CHECK = []
for it in range(10):
    c = np.random.normal(loc = Ec, scale = Dc, size = m)
    C = np.sum(c)
    C2 = C**2
    CHECK.append(C2)
    print(f"it{it}: {C}")

EC2 = (m*Dc) + (m*Ec)**2
print(np.mean(CHECK))
print(EC2)
```

```python
# formular of EП
m = 100

Ec = 2
Dc = 0.5
sigma = 1

EC = m*Ec
EC2 = m*Dc + (m*Ec)**2
Ec2 = Dc + (Ec)**2

k = 2+m*sigma
EП_plus = 1/(8*k**2)*(4+sigma**2 * EC2 + k**2 * Ec2 + 4*sigma * EC - 2*sigma*k*Ec*EC - 4*k*Ec)
print(EП_plus)


CHECK = []
for it in range(1000):
    c = np.random.normal(loc = Ec, scale = Dc**0.5, size = m)
    i = random.randint(0, m-1)
    EP = 1/8*((2+sigma*np.sum(c))/(2+m*sigma)-c[i])**2
    CHECK.append(EP)
    #print(f"iteration{it}: {EP}")
print(np.mean(CHECK))
```
